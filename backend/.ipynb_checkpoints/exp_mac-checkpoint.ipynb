### Import the chat.txt
> 1. Read the chat form the chat folder
> 2. define the regex `pattern` for seperating `Date_Time`, `User`, `Messages`

```python
with open('./chats/_chat_mac.txt') as f:
    content = f.read()
    pattern = r'\[(.*?)\] (.*?): (.*)'
    matches = re.findall(pattern, content)
    df = pd.DataFrame(matches, columns=["Date_Time", "User", "Message"])

df.head()
```

#### Split the `Date_Time` column into date and time in order to have some more flexiblity

For now just drop the `Time` column.

```python
# fix the data and time
df[['Date', 'Time']] = df.Date_Time.str.split(',', expand=True)
df.drop(["Date_Time", "Time"], axis=1, inplace=True)
df.head()
```

```python
# df['Date'] = pd.to_datetime(df['Date'], format="%dd/%mm/%Y")

df['Date'] = pd.to_datetime(df['Date'], format="mixed")
df.head()
```

```python
import spacy
 
# Load spaCy English model
nlp = spacy.load("en_core_web_sm")
 
# Sample text
text = "There is a pen on the table"
 
# Process the text using spaCy
doc = nlp(text)
 
# Remove stopwords
filtered_words = [token.text for token in doc if not token.is_stop]
 
# Join the filtered words to form a clean text
clean_text = ' '.join(filtered_words)
 
print("Original Text:", text)
print("Text after Stopword Removal:", clean_text)
```
